```python
# Regresja liniowa: Najprostszy model, zakłada liniową relację między zmiennymi.
# Modele nieliniowe: Używane, gdy relacja między zmiennymi nie jest liniowa (np. funkcje wykładnicze, logarytmiczne).
# Modele wielomianowe: Przydatne, gdy dane są bardziej skomplikowane, więcej niż jedna zmienna niezależna
# Modele oparte na równaniach różniczkowych: Często stosowane w biotechnologii do opisu dynamiki procesów, np. wzrostu mikroorganizmów
```

```python
# Przykład 1


import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# zawarość biomasy (mikroorganizmów), sucha masa, mg
x_data = np.array([0, 10, 33, 45, 56])
# mętność próbki (OD 600nm)
y_data = np.array([0, 0.2, 0.66, 0.91, 1.34])
plt.plot(x_data,y_data,'bo')
```

```python
# nasz model: ax+b (albo coś bardziej skomplikowanego)
def model(x, a, b):
    return a*x + b

# Dopasowanie funkcji do danych
popt, pcov = curve_fit(model, x_data, y_data)
a, b = popt

label = f"Dopasowane wartości parametrów: a = {a:.2f}, b = {b:.2f}"

# Przewidywanie wartości y na podstawie dopasowanego modelu
#x = linspace(0,50)
#y_fit = model(x, a, b)
y_fit = model(x_data, a, b)

plt.plot(x_data,y_data,'bo')
plt.plot(x_data, y_fit,'-', label=label)
plt.legend()
plt.xlabel("Ilość biomasy, mg d.w.")
plt.ylabel("OD (600nm), A.U.")
```

```python
#X (stężenie laktozy, kg/m³)
x_data = np.array([0,
              0.33, 
              0.8,
              1.4,
              2.0,
              3.2,
              3.9,
              4.7])

y_data = np.array([0,
              1e-5,
              3e-5,
              5e-5,
              5.5e-5,
              4.9e-5,
              2.9e-5,
              1.1e-5])

plt.plot(x_data,y_data, 'ko')


# nasz model: ax^2+bx+c
def model(x, a, b, c):
    return a*x**2 + b*x + c

# Dopasowanie funkcji do danych
popt, pcov = curve_fit(model, x_data, y_data)
a, b, c = popt

label = f"Dopasowanie: a = {a:.2e}, b = {b:.2e}, c = {c:.2e} "

# Przewidywanie wartości y na podstawie dopasowanego modelu
x = np.linspace(0,5)
y_fit = model(x, a, b, c)
#y_fit = model(x_data, a, b, c)

plt.plot(x_data,y_data,'bo')
plt.plot(x, y_fit,'-', label=label)
plt.legend()
plt.xlabel("Stężenie laktozy, $kg/m^3$")
plt.ylabel("Szybkość właściwa fermentacji, $\mu$ , 1/s")
```
